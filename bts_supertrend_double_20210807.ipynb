
## CUSTOM INDICATOR - SUPERTREND


```python

class SuperTrendBand(bt.Indicator):
    """
    Helper inidcator for Supertrend indicator
    """
    params = (('period',RUN_PARAMS['STRAT_PARAMS']['PERIOD']),('multiplier',RUN_PARAMS['STRAT_PARAMS']['MULTI']))
    lines = ('basic_ub','basic_lb','final_ub','final_lb')


    def __init__(self):
        self.atr = bt.indicators.AverageTrueRange(period=self.p.period)
        self.l.basic_ub = ((self.data.high + self.data.low) / 2) + (self.atr * self.p.multiplier)
        self.l.basic_lb = ((self.data.high + self.data.low) / 2) - (self.atr * self.p.multiplier)

    def next(self):
        if len(self)-1 == self.p.period:
            self.l.final_ub[0] = self.l.basic_ub[0]
            self.l.final_lb[0] = self.l.basic_lb[0]
        else:
            #=IF(OR(basic_ub<final_ub*,close*>final_ub*),basic_ub,final_ub*)
            if self.l.basic_ub[0] < self.l.final_ub[-1] or self.data.close[-1] > self.l.final_ub[-1]:
                self.l.final_ub[0] = self.l.basic_ub[0]
            else:
                self.l.final_ub[0] = self.l.final_ub[-1]

            #=IF(OR(baisc_lb > final_lb *, close * < final_lb *), basic_lb *, final_lb *)
            if self.l.basic_lb[0] > self.l.final_lb[-1] or self.data.close[-1] < self.l.final_lb[-1]:
                self.l.final_lb[0] = self.l.basic_lb[0]
            else:
                self.l.final_lb[0] = self.l.final_lb[-1]

class SuperTrend(bt.Indicator):
    global STRAT_PARAMS
    """
    Super Trend indicator
    """
    params = (('period', STRAT_PARAMS['PERIOD']), ('multiplier', STRAT_PARAMS['MULTI']))
    lines = ('super_trend',)
    plotinfo = dict(subplot=False)

    def __init__(self):
        self.stb = SuperTrendBand(period = self.p.period, multiplier = self.p.multiplier)

    def next(self):
        if len(self) - 1 == self.p.period:
            self.l.super_trend[0] = self.stb.final_ub[0]
            return

        if self.l.super_trend[-1] == self.stb.final_ub[-1]:
            if self.data.close[0] <= self.stb.final_ub[0]:
                self.l.super_trend[0] = self.stb.final_ub[0]
            else:
                self.l.super_trend[0] = self.stb.final_lb[0]

        if self.l.super_trend[-1] == self.stb.final_lb[-1]:
            if self.data.close[0] >= self.stb.final_lb[0]:
                self.l.super_trend[0] = self.stb.final_lb[0]
            else:
                self.l.super_trend[0] = self.stb.final_ub[0]
                
if RUN_PARAMS['DEBUG']:
    print(datetime.now())
```

```python
# Create a Stratey

class supertrend_double(bt.Strategy):
    global DEBUG, BUYSELLS_THIS_ALGO, FIRST_CLOSE, FIRST_NOTIFY_TRADE
    global algodicators, algofilename, algofilepath
    lines = ('TREND','TREND_FLIP', 'PCT_CHG_CLOSE')

    FIRST_NOTIFY_TRADE = True
    
    def log(self, txt, dt=None):
        ''' Logging function fot this strategy'''
        #dt = dt or self.datas[0].datetime.date(0)
        dt = dt or self.datas[0].datetime.datetime()
        
        if RUN_PARAMS['DEBUG']:
            print('%s @    %s' % (dt.strftime("%Y-%m-%d %H:%M"), txt))

    def __init__(self):
        # Keep a reference to the "close" line in the data[0] dataseries
        
        
        
        # OHLC INFORMATION
        #self.open = self.datas.lines.open[0] # shorthand below
        self.open = self.datas[0].lines
        
        #self.high = self.datas.lines.high[0] # shorthand below        
        self.high = self.datas[0].high
        
        #self.low = self.datas.lines.low[0] # shorthand below                
        self.low = self.datas[0].low        
        
        #self.close = self.datas.lines.close[0] # shorthand below                
        self.close = self.datas[0].close   
        self.close_mean = round(np.mean(self.data.close),4)
        
        self.ORDER_IN_PROGRESS_TYPE = 'START'

        # FOR SAVING FILES below 
        #self.firstrow = True
        #self.SAVE_ROW = True

        # To keep track of pending orders
        self.order = None
        
        # To get the first close        
        self.FIRST_CLOSE = 0 
        
        # Tracking buys and sells
        self.COUNT_BUY = 0 
        self.COUNT_SELL = 0 
        self.TAKE_PROFIT = 0
        self.STOP_LOSS = 0
        
        self.stb = SuperTrendBand() # lines = ('basic_ub','basic_lb','final_ub','final_lb')
        self.stb.plotinfo.plot = True
        self.stb.csv = True
        
        self.supertrend = SuperTrend() # lines = ('super_trend',)
        self.supertrend.plotinfo.plot = True
        self.supertrend.csv = True
                
        # defined as lines, instead of attributes, can be initialised in nextstart()
        #self.TREND = False 
        #self.TREND_FLIP = False                
        #self.pct_chg_close = 0
        
    def nextstart(self):# Creating files for saving         
        global algoid, algofilename, csvid, algofilepath
        
        self.FIRST_CLOSE = self.close
        self.PCT_CHG_CLOSE[0] = 0
        
        # Keeping TREND = False implies that buy order won't be created immidiately
        self.TREND[0] = False        
        self.TREND_FLIP[0] = False
        
        algoid = datetime.now().strftime("%Y_%m_%d_%H_%M")
        csvid = datetime.now().strftime("%Y_%m_%d_%H") # Influences the name of file where we save the SELL and BUY Data        
        
    def notify_trade(self, trade):
        global DEBUG, FIRST_NOTIFY_TRADE
        #self.log('')
        print()
        
        if FIRST_NOTIFY_TRADE and DEBUG:
            self.order_status_dict = {'Created' : self.order.Created, 'Submitted' : self.order.Submitted, 'Accepted' : self.order.Accepted,  'Partial' : self.order.Partial ,  'Completed' : self.order.Completed \
                                      , 'Cancelled' : self.order.Cancelled ,'Expired':self.order.Expired, 'Margin':self.order.Margin ,  'Rejected' : self.order.Rejected }
            self.log(f'nt: order_status_dic {self.order_status_dict}')
            FIRST_NOTIFY_TRADE = False
            
        
        # SAVE_ROW logic to avoid save this row thrice into file
        # if DEBUG:
        self.log('nt: notify_trade() executed' )
        self.log(f'nt: self.order.status : {self.order.status}' )
                
        if DEBUG:
            pass
            #self.log(f'nt: order status is {self.order.status}')
            #self.log(f'nt: submitted status is {self.order.Submitted}')
            #self.log(f'nt: Accepted status is {self.order.Accepted}')
            #self.log(f'nt: Completed status is {self.order.Completed}')
                
        
        if self.order.status in [self.order.Submitted, self.order.Accepted]:
            # Buy/Sell order submitted/accepted to/by broker - Nothing to do
            return

        # Check if an order has been completed
        # Attention: broker could reject order if not enough cash
        if self.order.status in [self.order.Completed]:
            if self.order.isbuy():
                self.log('nt: BUY EXECUTED, %.4f' % self.order.executed.price)
            elif self.order.issell():
                self.log('nt: SELL EXECUTED, %.4f' % self.order.executed.price)

            self.bar_executed = len(self)

        elif self.order.status in [self.order.Canceled, self.order.Margin, self.order.Rejected]:
            self.log('nt: Order Canceled/Margin/Rejected')

        #self.save() # remnants of MANUAL saving mechanism
        
        # Printing trade.pnl and pnl.comm
        if trade.isopen:
            #self.trade_number += 1
            #print('Opening trade #%02d with size %.4f commision %.4f value %.2f price %.2f' % (self.trade_number, trade.size, trade.commission, trade.value, trade.price))
            self.log('nt: Opening trade with size %.4f commision %.4f value %.2f price %.2f' % (trade.size, trade.commission, trade.value, trade.price))
        elif trade.isclosed:
            #print('Closing trade #%02d gross %.2f, net %.2f' %(self.trade_number, trade.pnl, trade.pnlcomm))
            self.log('nt: Closing trade gross %.2f, net %.2f' %(trade.pnl, trade.pnlcomm))
        
        # After order placed
        self.log(f'nt: self.order.status : {self.order.status}' )
        
        # Write down: no pending order
        self.order = None
        #self.log('\n')
        print()

    def next(self):
        self.log('next :')
        self.log(f'{type(self.order)}')
        self.log(f'{self.position}')
        
        if self.position:
            self.log('Position is True')
        elif not self.position:
            self.log('Position is False')

        # Simply log the closing price of the series from the reference
        #if DEBUG: self.log('Close, %.4f' % self.close[0])
        
        #self.log(f'self.order.status : {self.order.status}')
        
        self.PCT_CHG_CLOSE[0] = self.close[0] / self.close[-1] - 1
        #if self.firstrow:
        #    self.firstrow = False
        #else:
        #    self.PCT_CHG_CLOSE[0] = self.close[0] / self.close[-1] - 1
        
        # Check if an order is pending ... if yes, we cannot send a 2nd one
        if self.order:
            return

        if self.close > self.stb.final_ub[-1]:
            self.TREND[0] = True
        elif self.close < self.stb.final_lb[-1]:
            self.TREND[0] = False
        else:
            self.TREND[0] = self.TREND[-1]

            if self.TREND[0] and self.stb.final_lb[0] < self.stb.final_lb[-1]:
                self.stb.final_lb[0] = self.stb.final_lb[-1]

            if not self.TREND[0] and self.stb.final_ub[0] > self.stb.final_ub[-1]:
                self.stb.final_ub[0] = self.stb.final_ub[-1]
            
        if self.TREND[0] == self.TREND[-1]:
            self.TREND_FLIP[0] = False
        elif self.TREND[0] != self.TREND[-1]:
            self.TREND_FLIP[0] = True
            
        # Check position for creating buy or sell signals
        
        # DOUBLING DOWN ON BUYS AND SELLS
        if not self.position:
            if self.TREND_FLIP[0]:
                # IF trend flips check to see what he trend indicates BUY or SELL
                if self.TREND[0]:
                    self.log('next: MAIN BUY')
                    self.order = self.buy(size = STRAT_PARAMS['TRADE_SIZE'] , exectype = bt.Order.Market)                
                    self.ORDER_IN_PROGRESS_TYPE = 'BUY'
                    self.COUNT_BUY = self.COUNT_BUY + 1
                if (not self.TREND[0]):
                    self.log('next: MAIN SELL')
                    self.order = self.sell(size = STRAT_PARAMS['TRADE_SIZE'] , exectype = bt.Order.Market)
                    self.ORDER_IN_PROGRESS_TYPE = 'SELL'
                    self.COUNT_SELL = self.COUNT_SELL + 1
                
        if self.position:
            
            if self.ORDER_IN_PROGRESS_TYPE == 'SELL':
                if self.TREND_FLIP[0] and self.TREND[0]:
                    self.log('next: SECONDARY BUY')
                    self.log(f'{type(self.order)}')
                    self.order = self.buy(size = STRAT_PARAMS['TRADE_SIZE'], exectype = bt.Order.Market)
                    self.log(f'{type(self.order)}')
                    #self.order = self.buy(size = STRAT_PARAMS['TRADE_SIZE'] + STRAT_PARAMS['TRADE_SIZE'], exectype = bt.Order.Market) # buy to close SELL order and place ANOTHER BUY order    
                    self.ORDER_IN_PROGRESS_TYPE = 'BUY'
                    self.COUNT_BUY = self.COUNT_BUY + 1
                    
            if self.ORDER_IN_PROGRESS_TYPE == 'BUY':
                if self.TREND_FLIP[0] and (not self.TREND[0]):
                    self.log('next: SECONDARY SELL')
                    self.log(f'{type(self.order)}')
                    self.order = self.sell(size = STRAT_PARAMS['TRADE_SIZE'], exectype = bt.Order.Market)
                    self.log(f'{type(self.order)}')
                    #self.order = self.sell(size = STRAT_PARAMS['TRADE_SIZE'] + STRAT_PARAMS['TRADE_SIZE'], exectype = bt.Order.Market) # sell to close BUY order and place ANOTHER SELL order
                    self.ORDER_IN_PROGRESS_TYPE = 'SELL'
                    self.COUNT_SELL = self.COUNT_SELL + 1

                
    def stop(self):        
        global algoid, algofilename
        global BUYSELLS_THIS_ALGO, FIRST_CLOSE
        global ALGO_PERF
        

        #ALGO_PERF = {}
        
        portfolio = cerebro.broker.getvalue()
        cash = 0 
        profit_value = round(portfolio - STRAT_PARAMS['ACCOUNT_BALANCE'],4)
        profit_pct_on_INIBAL = round( ((portfolio + cash - STRAT_PARAMS['ACCOUNT_BALANCE']) / STRAT_PARAMS['ACCOUNT_BALANCE'])*100,4)
        
        #profit_pct_on_first_close = ((portfolio + cash - STRAT_PARAMS['ACCOUNT_BALANCE']) / STRAT_PARAMS['FIRST_CLOSE'])*100
        profit_pct_on_first_close = round( ((portfolio + cash - STRAT_PARAMS['ACCOUNT_BALANCE']) / self.FIRST_CLOSE)*100, 4)
        profit_pct_on_close_mean = round( ((portfolio + cash - STRAT_PARAMS['ACCOUNT_BALANCE']) / self.close_mean)*100, 4)
       
        #print(f"stop: Profit {profit_value:,.2f}$")
        #print(f"stop: Profit {profit_pct_on_INIBAL:.2f} % on BAL ")
        #print(f"stop: Profit {profit_pct_on_close_mean:.2f} % on MEAN OF CLOSING PRICES of {self.close_mean}")
        
        BUYSELLS_THIS_ALGO = self.COUNT_BUY + self.COUNT_SELL

        ALGO_PERF = {'algoid':algoid, 'algoname':STRATEGY_NAME, 'pair':cryptopair_, 'kline':kline_, 'start_dt' : start_date_, 'end_dt' : end_date_, 'params' : STRAT_PARAMS, 'pft_val':profit_value, 'close_lst':self.close[0] ,'close_mean':self.close_mean ,'pft_pct_on_mean':profit_pct_on_close_mean \
                     , 'BUYS': self.COUNT_BUY, 'SELLS':self.COUNT_SELL}
        #print(ALGO_PERF)        

if RUN_PARAMS['DEBUG']:
    print(datetime.now())
```
