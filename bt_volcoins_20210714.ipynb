```python


#from strategies import TestStrategy
#from strategies import strat_bb_based
#from strategies import GoldenCross

# Create a Stratey

class volatility(bt.Strategy):
    global DEBUG, BUYSELLS_THIS_ALGO, FIRST_CLOSE
    global algodicators, algofilename, algofilepath

    def log(self, txt, dt=None):
        ''' Logging function fot this strategy'''
        #dt = dt or self.datas[0].datetime.date(0)
        dt = dt or self.datas[0].datetime.datetime()
        print('%s @    %s' % (dt.strftime("%Y-%m-%d %H:%M"), txt))

    def __init__(self):
        # Keep a reference to the "close" line in the data[0] dataseries
        self.open = self.datas[0].open
        self.high = self.datas[0].high
        self.low = self.datas[0].low
        self.close = self.datas[0].close   
        
        self.ORDER_IN_PROGRESS_TYPE = 'START'
        
        # FOR SAVING FILES below 
        self.firstrow = True
        self.SAVE_ROW = True

        # To keep track of pending orders
        self.order = None
        
        # To get the first close        
        self.FIRST_CLOSE = 0 
        
        # Tracking buys and sells
        self.COUNT_BUY = 0 
        self.COUNT_SELL = 0 
        self.TAKE_PROFIT = 0
        self.STOP_LOSS = 0
        
    def save(self):
        global algoid, DEBUG, csvid
        dt = self.datas[0].datetime.datetime()
        
        open_ = self.open[0]
        high_ = self.high[0]
        low_ = self.low[0]
        close_ = self.close[0]
        
        pf_val = self.broker.getvalue()
        cash = self.broker.get_cash()
        
        possize = self.position.size
        #possize = self.getposition.size
        #pnl = round(trade_arg.pnl, 4)
        
        row = [ algoid, dt, STRATEGY_NAME, open_, high_, low_, close_, self.ORDER_IN_PROGRESS_TYPE, pf_val, cash, possize]
        self.log( 'save: Writing row {}'.format(row))
        #self.log('\n')
                    
        #with open('{}\\{}_{}_{}_{}.csv'.format(algodicators, STRATEGY_NAME, algoid, cryptopair_, kline_), 'a', newline='') as f:
        with open('{}\\{}_{}_{}_{}.csv'.format(algodicators, STRATEGY_NAME, csvid, cryptopair_, kline_), 'a', newline='') as f:
            csv.writer(f, dialect = 'excel').writerow(row)
        f.close()

                
    def nextstart(self):# Creating files for saving         
        global algoid, algofilename, csvid, algofilepath
        
        algoid = datetime.now().strftime("%Y_%m_%d_%H_%M")
        csvid = datetime.now().strftime("%Y_%m_%d_%H") # Influences the name of file where we save the SELL and BUY Data
        
        #fileexists = os.path.isfile('{}\\{}_{}_{}_{}.csv'.format(algodicators, STRATEGY_NAME, algoid, cryptopair_, kline_))
        fileexists = os.path.isfile('{}\\{}_{}_{}_{}.csv'.format(algodicators, STRATEGY_NAME, csvid, cryptopair_, kline_))

        if fileexists:
            pass
        else:
            headers = ['algoid','datetime', 'strategy', 'open', 'high', 'low','close', 'order', 'portfolio', 'cash', 'possize']
            
            #with open('{}\\{}_{}_{}_{}.csv'.format(algodicators, STRATEGY_NAME, algoid, cryptopair_, kline_), 'a', newline='') as f:
            with open('{}\\{}_{}_{}_{}.csv'.format(algodicators, STRATEGY_NAME, csvid, cryptopair_, kline_), 'a', newline='') as f:
                csv.writer(f, dialect = 'excel').writerow(headers)
            f.close()
            
            self.log( 'nextstart: Writing headers {}'.format(headers))
            self.log('\n')
        self.save()
        
        algofilename = 'Algo_{}_{}'.format(date.today().strftime("%Y_%m_%d_%H"),RUN_TYPE)
        algofilepath = os.path.isfile('{}\\{}.csv'.format(algodicators, algofilename))
        
        if not algofilepath:
            # write headers
            algoheaders = [ 'algoid', 'algo', 'cryptopair', 'kline', 'start_date', 'end_date', 'params', 'profit_pct_on_INIBAL', 'profit_pct_on_first_close', 'first_close', 'count_buy', 'count_sell', 'other_performance_params']
            with open('{}\\{}.csv'.format(algodicators, algofilename), 'a', newline='') as f:
                csv.writer(f, dialect = 'excel').writerow(algoheaders)
            f.close()
            self.log( 'nextstart: Writing headers {}'.format(algoheaders))
                        
            self.log('\n')
                
    def notify_trade(self, trade):
        global DEBUG
        # ERROR BELOW
        # self.log(self.order.executed.value)
        
        # SAVE_ROW logic to avoid save this row thrice into file
        if DEBUG:
            self.log('nt: notify_trade() executed' )
                
        if DEBUG:
            pass
            #self.log(f'nt: order status is {self.order.status}')
            #self.log(f'nt: submitted status is {self.order.Submitted}')
            #self.log(f'nt: Accepted status is {self.order.Accepted}')
            #self.log(f'nt: Completed status is {self.order.Completed}')
                
        
        if self.order.status in [self.order.Submitted, self.order.Accepted]:
            # Buy/Sell order submitted/accepted to/by broker - Nothing to do
            return

        # Check if an order has been completed
        # Attention: broker could reject order if not enough cash
        if self.order.status in [self.order.Completed]:
            if self.order.isbuy():
                self.log('nt: BUY EXECUTED, %.4f' % self.order.executed.price)
            elif self.order.issell():
                self.log('nt: SELL EXECUTED, %.4f' % self.order.executed.price)

            self.bar_executed = len(self)

        elif self.order.status in [self.order.Canceled, self.order.Margin, self.order.Rejected]:
            self.log('nt: Order Canceled/Margin/Rejected')

        self.save()
        
        # Write down: no pending order
        self.order = None
        self.log('\n')

    def next(self):
        
        # SAVE_ROW makes sure we save every kind of order 
        #self.SAVE_ROW = True
        
        if self.firstrow:
            self.FIRST_CLOSE = self.close[0]
            self.firstrow = False
        
        # Simply log the closing price of the series from the reference
        #if DEBUG: self.log('Close, %.4f' % self.close[0])

        # Check if an order is pending ... if yes, we cannot send a 2nd one
        if self.order:
            return

        # Check if we are in the market
        if not self.position:

            # open a position if the current candle price is higher by at least 1%
            # compared to the previous candle
            if STRAT_PARAMS['CHANGE_IN_PRICE'] > 0:
                BUY_CONDITION = self.close[0] >= self.close[-1] + (self.close[-1] * STRAT_PARAMS['CHANGE_IN_PRICE']/100)
            elif STRAT_PARAMS['CHANGE_IN_PRICE'] < 0:
                BUY_CONDITION = self.close[0] <= self.close[-1] + (self.close[-1] * STRAT_PARAMS['CHANGE_IN_PRICE']/100)
            else:
                BUY_CONDITION = None
                                
            if BUY_CONDITION: 
                    self.log('next: BUY CREATE, %.4f' % self.close[0])
                    self.COUNT_BUY = self.COUNT_BUY + 1 

                    # Keep track of the created order to avoid a 2nd order
                    self.buy(size = STRAT_PARAMS['TRADE_SIZE'], exectype=bt.Order.Market)
                    self.order = self.buy()
                    self.ORDER_IN_PROGRESS_TYPE = 'BUY'

                    self.log(f'next: EXECUTED PRICE IS:{self.order.executed.price}')

        if self.position:
            if self.close[0] > self.position.price + (self.position.price * STRAT_PARAMS['TAKE_PROFIT']/100): 
                # Already in the market ... we might sell
                self.log('next: SELL CREATE - TAKE_PROFIT, %.4f' % self.close[0])
                self.sell(size = STRAT_PARAMS['TRADE_SIZE'], exectype = bt.Order.Market)
                self.order = self.sell()
                self.ORDER_IN_PROGRESS_TYPE = 'CLOSE'
                
                self.COUNT_SELL = self.COUNT_SELL + 1 
                self.TAKE_PROFIT = self.TAKE_PROFIT + 1 
                
            elif self.close[0] < self.position.price - (self.position.price * STRAT_PARAMS['STOP_LOSS']/100):
                # Already in the market ... we might sell
                self.log('next: SELL CREATE - STOP_LOSS, %.4f' % self.close[0])
                self.sell(size = STRAT_PARAMS['TRADE_SIZE'], exectype = bt.Order.Market)
                self.order = self.sell()
                self.ORDER_IN_PROGRESS_TYPE = 'CLOSE'
                
                self.COUNT_SELL = self.COUNT_SELL + 1 
                self.STOP_LOSS = self.STOP_LOSS + 1 

                
    def stop(self):        
        global algoid, algofilename
        global BUYSELLS_THIS_ALGO, FIRST_CLOSE
        algofilepath = os.path.isfile('{}\\{}.csv'.format(algodicators, algofilename))
        
        portfolio = cerebro.broker.getvalue()
        cash = 0 
        profit_value = round(portfolio - STRAT_PARAMS['ACCOUNT_BALANCE'],4)
        profit_pct_on_INIBAL = round( ((portfolio + cash - STRAT_PARAMS['ACCOUNT_BALANCE']) / STRAT_PARAMS['ACCOUNT_BALANCE'])*100,4)
        
        #profit_pct_on_first_close = ((portfolio + cash - STRAT_PARAMS['ACCOUNT_BALANCE']) / STRAT_PARAMS['FIRST_CLOSE'])*100
        profit_pct_on_first_close = round( ((portfolio + cash - STRAT_PARAMS['ACCOUNT_BALANCE']) / self.FIRST_CLOSE)*100, 4)
        
        print(f"stop: Profit {profit_value:,.2f}$")
        print(f"stop: Profit {profit_pct_on_INIBAL:.2f} % on BAL ")

        print(f"stop: Profit {profit_pct_on_first_close:.2f} % on FIST CLOSING PRICE ")
        
        if algofilepath:
            # write data if file exists already
            algoheaders = [ 'algoid', 'algo', 'cryptopair', 'kline', 'start_date', 'end_date', 'params', 'profit_pct_on_INIBAL', 'profit_pct_on_first_close', 'first_close', 'count_buy', 'count_sell', 'other_performance_params']
            algodetails = [ algoid, STRATEGY_NAME, cryptopair_, kline_, start_date_, end_date_, STRAT_PARAMS, profit_pct_on_INIBAL, profit_pct_on_first_close, self.FIRST_CLOSE, self.COUNT_BUY, self.COUNT_SELL, {'TAKE_PROFIT':self.TAKE_PROFIT, 'STOP_LOSS':self.STOP_LOSS}]
            with open('{}\\{}.csv'.format(algodicators, algofilename), 'a', newline='') as f:
                csv.writer(f, dialect = 'excel').writerow(algodetails)
            f.close()
            
        BUYSELLS_THIS_ALGO = self.COUNT_BUY + self.COUNT_SELL
        FIRST_CLOSE = self.FIRST_CLOSE
             


print(datetime.now())
```
